```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("inequality_output.csv")
data.head()
```

```python
final = data[data["Step"]==5]
final = list(final["Wealth"])
```

```python
plt.hist(final, bins = 100)
None
```

```python
X= list(range(1, 2001))
y = sorted(final)
```

```python
plt.scatter(X,y)
```

```python
ccdf =[]
for idx in range(len(final)): 
    if len(ccdf) == 0: 
        ccdf.append(final[idx])
    else: 
        ccdf.append(final[idx]+ccdf[-1])
```

```python
plt.plot(X,ccdf)
```
